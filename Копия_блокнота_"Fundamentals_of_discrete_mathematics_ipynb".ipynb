# ЗАВДАННЯ 2: БУЛЕВА АЛГЕБРА
## Контроль доступу (Логічні операції і таблиця істинності)

```python
from itertools import product

print("\nЗАВДАННЯ 2: КОНТРОЛЬ ДОСТУПУ")

print("\nФормули доступу:")
print("B₀ (Base) = E ∧ V ∧ ¬B")
print("B₁ (Premium) = (E ∨ P) ∧ V ∧ ¬B")
print("B₂ (Admin) = A ∧ V ∧ ¬B")
print("B₃ (Secret) = (A ∨ (E ∧ P)) ∧ V ∧ ¬B")
print("\nПараметри: E - employee, V - verified, P - premium, A - admin, B - banned")
```

```python
# Функція для обчислення доступу
def check_access(E, V, P, A, B):
    B0 = E and V and not B
    B1 = (E or P) and V and not B
    B2 = A and V and not B
    B3 = (A or (E and P)) and V and not B
    return B0, B1, B2, B3

# Генеруємо всі 32 комбінації
print("\n2.1) Таблиця істинності (32 комбінації)")
print("\nE V P A B | B₀ B₁ B₂ B₃")
print("-" * 35)

all_results = []
for E, V, P, A, B in product([0, 1], repeat=5):
    b0, b1, b2, b3 = check_access(E, V, P, A, B)
    all_results.append({
        'E': E, 'V': V, 'P': P, 'A': A, 'B': B,
        'B0': int(b0), 'B1': int(b1), 'B2': int(b2), 'B3': int(b3)
    })
    print(f"{E} {V} {P} {A} {B} | {int(b0)} {int(b1)} {int(b2)} {int(b3)}")
```

```python
from itertools import product

# Функція для обчислення доступу
def check_access(E, V, P, A, B):
    B0 = E and V and not B
    B1 = (E or P) and V and not B
    B2 = A and V and not B
    B3 = (A or (E and P)) and V and not B
    return B0, B1, B2, B3

# Генеруємо всі 32 комбінації
all_results = []
for E, V, P, A, B in product([0, 1], repeat=5):
    b0, b1, b2, b3 = check_access(E, V, P, A, B)
    all_results.append({
        'E': E, 'V': V, 'P': P, 'A': A, 'B': B,
        'B0': int(b0), 'B1': int(b1), 'B2': int(b2), 'B3': int(b3)
    })

# Питання 1: Повний доступ
print("\n2.2) Питання 1: Повний доступ до всіх 4 рівнів одночасно")

full_access = [r for r in all_results if r['B0'] and r['B1'] and r['B2'] and r['B3']]

print(f"\nКількість комбінацій з повmним доступом: {len(full_access)}")
print(f"\nВідповідь: {len(full_access)} комбінацій")
print(f"\nПояснення: Неможливо отримати повний доступ!")
print(f"Причина: Умови B₂ (A=1) та B₃ суперечливі з B₀ (E=1 обов'язково)")
```

```python
# Питання 2: Premium без Base
print("\n2.3) Питання 2: Чи існує Premium (B₁=1) без Base (B₀=0)?")

premium_without_base = [r for r in all_results if r['B1'] and not r['B0']]

print(f"\nКількість таких комбінацій: {len(premium_without_base)}")

if premium_without_base:
    print(f"\nВІДПОВІДЬ: ТА, існує! ({len(premium_without_base)} комбінацій)")
    ex = premium_without_base[0]
    print(f"\nПример: E={ex['E']}, V={ex['V']}, P={ex['P']}, A={ex['A']}, B={ex['B']}")
    print(f"\nПояснення:")
    print(f"  B₀ = {ex['E']} ∧ {ex['V']} ∧ {not ex['B']} = {ex['B0']} (НІ)")
    print(f"  B₁ = ({ex['E']} ∨ {ex['P']}) ∧ {ex['V']} ∧ {not ex['B']} = {ex['B1']} (ТАК)")
    print(f"\n  Якщо користувач НЕ працівник (E=0), але має преміум (P=1),")
    print(f"  то доступ до преміум рівня можливий, а до базового - НІ!")
else:
    print(f"\nВІДПОВІДЬ: НІ, не існує")

```

# ЗАВДАННЯ 3: КОМБІНАТОРИКА
## Команда мрія (Правило множення та сполучення)

```python
import math

print("\nЗАВДАННЯ 3: КОМАНДА МРІЯ")

print("\nДано:")
print("  Back-end розробників: 8")
print("  Front-end розробників: 6")
print("  Дизайнерів: 4")
print("\nПотрібно вибрати:")
print("  2 Back-end розробників")
print("  2 Front-end розробників")
print("  1 Дизайнера")

print("\n3.1) Обчислення сполучень")
```

```python
# Обчислюємо сполучення
n1, k1 = 8, 2  # Back-end
n2, k2 = 6, 2  # Front-end
n3, k3 = 4, 1  # Дизайнер

C_be = math.comb(n1, k1)
C_fe = math.comb(n2, k2)
C_des = math.comb(n3, k3)

print("\nФормула сполучень: C(n,k) = n! / (k! × (n-k)!)")

print(f"\nBack-end:")
print(f"  C(8,2) = 8! / (2! × 6!) = (8 × 7) / (2 × 1) = {C_be}")

print(f"\nFront-end:")
print(f"  C(6,2) = 6! / (2! × 4!) = (6 × 5) / (2 × 1) = {C_fe}")

print(f"\nДизайнер:")
print(f"  C(4,1) = 4! / (1! × 3!) = 4 / 1 = {C_des}")
```

```python
import math

# Обчислюємо сполучення (re-calculating for independent execution)
n1, k1 = 8, 2  # Back-end
n2, k2 = 6, 2  # Front-end
n3, k3 = 4, 1  # Дизайнер

C_be = math.comb(n1, k1)
C_fe = math.comb(n2, k2)
C_des = math.comb(n3, k3)

# Правило множення
print("\n3.2) Правило множення")

print(f"\nЯкщо першу подію можна здійснити n₁ способами,")
print(f"а другу - n₂ способами, то обидві можна здійснити")
print(f"n₁ × n₂ способами.")

print(f"\nОбщая формула:")
print(f"N = C(8,2) × C(6,2) × C(4,1)")
print(f"N = {C_be} × {C_fe} × {C_des}")

N = C_be * C_fe * C_des

print(f"\nN = {N} способів")

print(f"\nВідповідь: Можна сформувати {N} різних команд!")
```

# ЗАВДАННЯ 4: ТЕОРІЯ ГРАФІВ
## Соціальна мережа (Аналіз графа і теорема про суму степенів)

```python
# Граф спілкування
graph = {
    'Анна': ['Богдан', 'Віктор', 'Ганна'],
    'Богдан': ['Анна', 'Віктор', 'Дмитро'],
    'Віктор': ['Анна', 'Богдан', 'Ганна', 'Дмитро'],
    'Ганна': ['Анна', 'Віктор', 'Євген'],
    'Дмитро': ['Богдан', 'Віктор', 'Євген'],
    'Євген': ['Ганна', 'Дмитро']
}

# Знаходимо унікальні ребра
edges = set()
for person, contacts in graph.items():
    for contact in contacts:
        edge = tuple(sorted([person, contact]))
        edges.add(edge)

edges_list = sorted(edges)

print("\n4.2) Список ребер (зв'язків)")
print(f"\nВсього ребер: |E| = {len(edges_list)}\n")
for i, edge in enumerate(edges_list, 1):
    print(f"  {i:2}. ({edge[0]:8} — {edge[1]:8})")
```

```python
# Обчислюємо степені вершин
degrees = {person: len(contacts) for person, contacts in graph.items()}

print("\n4.3) Степені вершин (кількість контактів)")
print("\nСтепінь вершини = кількість ребер інцидентних вершині")

for person in sorted(degrees.keys()):
    deg = degrees[person]
    print(f"  deg({person[0]:1}) = {deg}   (контакти: {', '.join(graph[person])})")

max_person = max(degrees, key=degrees.get)
min_person = min(degrees, key=degrees.get)

print(f"\nМаксимальний степінь: {max_person} (deg = {degrees[max_person]}) ")
print(f"Мінімальний степінь: {min_person} (deg = {degrees[min_person]}) ")
```

```python
# Теорема про суму степенів
print("\n4.4) Теорема про суму степенів (Handshaking Lemma)")

print("\nТеорема: Σ deg(v) = 2|E|")
print("Сума всіх степенів дорівнює подвійній кількості ребер")

sum_degrees = sum(degrees.values())
edge_count = len(edges_list)

print(f"\nЛіва частина (сума степенів):")
deg_list = []
for person in sorted(degrees.keys()):
    deg_list.append(str(degrees[person]))
    print(f"  deg({person[0]}) = {degrees[person]}")

print(f"\n  Σ = {' + '.join(deg_list)} = {sum_degrees}")

print(f"\nПрава частина (подвійна кількість ребер):")
print(f"  |E| = {edge_count}")
print(f"  2|E| = 2 × {edge_count} = {2 * edge_count}")

if sum_degrees == 2 * edge_count:
    print(f"{sum_degrees} = {2 * edge_count}")
else:
    print(f"\nПОМИЛКА! {sum_degrees} ≠ {2 * edge_count}")

```
