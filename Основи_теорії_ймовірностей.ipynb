```python
#  2
import math

#  1:  
print(" 1:   ( 3 )")
p1 = (8/12) * (7/11) * (6/10)
print(f"P = (8/12)  (7/11)  (6/10) = {p1}")
```

```python
#  2: 
print("\n 2: ")

c12_3 = math.factorial(12) // (math.factorial(3) * math.factorial(9))
c8_3 = math.factorial(8) // (math.factorial(3) * math.factorial(5))

p2 = c8_3 / c12_3
print(f"C(12,3) = {c12_3}")
print(f"C(8,3) = {c8_3}")
print(f"P = {c8_3}/{c12_3} = {p2}")
```

```python
#  3:  2  ( )
print("\n 3:   ( 2 )")

p_ssd = (8/12) * (7/11) * (4/10)
p_sds = (8/12) * (4/11) * (7/10)
p_dss = (4/12) * (8/11) * (7/10)

p3a = p_ssd + p_sds + p_dss
print(f"P() = {p_ssd}")
print(f"P() = {p_sds}")
print(f"P() = {p_dss}")
print(f": {p3a}")
```

```python
#  3:  2  ()
print("\n 3: ")

c8_2 = math.factorial(8) // (math.factorial(2) * math.factorial(6))
c4_1 = 4

p3b = (c8_2 * c4_1) / c12_3
print(f"C(8,2)  C(4,1) / C(12,3) = {c8_2}  {c4_1} / {c12_3} = {p3b}")
```

```python
#  3

#   
pa = 1 - 0.02
pb = 1 - 0.05
pc = 1 - 0.03
pd = 1 - 0.04

print(f"P(A ) = {pa}")
print(f"P(B ) = {pb}")
print(f"P(C ) = {pc}")
print(f"P(D ) = {pd}")
```

```python
#  1:  
print("\n 1:  ")
p_all = pa * pb * pc * pd
print(f"P = {pa}  {pb}  {pc}  {pd} = {p_all}")
```

```python
#  2:    
print("\n 2:    ")
p_fail = 1 - p_all
print(f"P = 1 - {p_all} = {p_fail}")
```

```python
#  3:  2 
print("\n 3:  2 ")

# 6 
ab = 0.02 * 0.05 * pc * pd
ac = 0.02 * pb * 0.03 * pd
ad = 0.02 * pb * pc * 0.04
bc = pa * 0.05 * 0.03 * pd
bd = pa * 0.05 * pc * 0.04
cd = pa * pb * 0.03 * 0.04

p_two = ab + ac + ad + bc + bd + cd

print(f"P(A,B) = {ab}")
print(f"P(A,C) = {ac}")
print(f"P(A,D) = {ad}")
print(f"P(B,C) = {bc}")
print(f"P(B,D) = {bd}")
print(f"P(C,D) = {cd}")
print(f": {p_two}")
```

```python
#  3-4:  bayes_source()
print("\n 3-4:  bayes_source()")

def bayes_source(prior_probs, conversion_rates):
    #  
    total = 0
    for i in range(len(prior_probs)):
        total += prior_probs[i] * conversion_rates[i]

    #  
    result = []
    for i in range(len(prior_probs)):
        p = (prior_probs[i] * conversion_rates[i]) / total
        result.append(p)

    return result

print("!")
```

```python
# 
priors = [0.50, 0.30, 0.20]
conversions = [0.04, 0.02, 0.08]

results = bayes_source(priors, conversions)

print(":")
print(f": {results[0]:.4f}")
print(f": {results[1]:.4f}")
print(f"Email: {results[2]:.4f}")
```
